```python
import os
csvs = [file for file in os.listdir() if file.endswith('csv') and not file.startswith('wait')]
```

```python
csvs
```

```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.DataFrame()

for csv in csvs:
    
    df = pd.read_csv(csv, header=None)
    df = df[1:]
    df.columns = ['attraction', 'datetime', 'wait']
    df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S')
    df['datetime'] = df['datetime'].apply(lambda x: datetime(x.year, x.month, x.day, x.hour, x.minute))
    df['time'] = df['datetime'].dt.time
    df['wait'] = df['wait'].astype(int)
    df.drop(columns=['datetime'], inplace=True)

    data = pd.concat([data, df], axis=0)
```

```python
import numpy as np
data['time'] = data['time'].astype(str)
data = data.groupby(['attraction', 'time'])['wait'].mean().reset_index()
data
```

```python
data = data.pivot(columns='attraction', values='wait', index='time')
data = data.loc['09:00:00':'21:00:00']
data
```

```python
data.to_csv('wait.csv')
```
